```python
import pandas as pd
```

```python
data = pd.read_csv(r'../data/HR.csv')
data.info()
```

```python
all_features = pd.concat([data.iloc[:,:6], data.iloc[:,7:]], axis=1)
all_labels = data.left
all_features.info()
```

```python
numeric_features = all_features.dtypes[all_features.dtypes != 'object'].index
numeric_features
```

```python
all_features[numeric_features] = all_features[numeric_features].apply(
    lambda x: (x - x.mean()) / (x.std())
)
all_features
```

```python
all_features = pd.get_dummies(all_features, dummy_na=True)
all_features
```

```python
import torch
from torch import nn
import torch.nn.functional as F
```

```python
X = torch.from_numpy(all_features.values).type(torch.float32)
Y = torch.from_numpy(all_labels.values.reshape(-1, 1)).type(torch.float32)
```

```python
in_features = X.shape[1]

class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear1 = nn.Linear(in_features, 32)
        self.linear2 = nn.Linear(32, 16)
        self.linear3 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        x = F.relu(self.linear1(x))
        x = F.relu(self.linear2(x))
        x = F.relu(self.linear3(x))
        x = self.sigmoid(x)
        return x
```

```python
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, shuffle=True)
X_train.shape, X_test.shape
```

```python
BATCHSIZE = 64
train_iter = DataLoader(TensorDataset(X_train, Y_train), batch_size=BATCHSIZE, shuffle=True)
test_iter = DataLoader(TensorDataset(X_test, Y_test), batch_size=BATCHSIZE)
```

```python
model = Net()
loss = nn.BCELoss()
opt = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 100

for epoch in range(num_epochs):
    # acc = 0
    # total = 0
    for x, y in train_iter:
        opt.zero_grad()
        l = loss(model(x), y)
        l.backward()
        opt.step()
    # with torch.no_grad():
    #     for x, y in test_iter:
    #         acc += ((model(x).data.numpy() > 0.5).astype('int') == y.numpy()).mean()
    #         total += 1
    # print(f'{epoch + 1} of {num_epochs} accuracy: {round((acc / total) * 100, 3)}')
```
